Puis, on a la fonction 'load' qui ne fait que charger le contenu du fichier JSON dans la variable « data ».

```python
import json
from pymongo import MongoClient
```

```python
def collection(uri):
    Client=MongoClient(uri)
    database=Client["rhobs"]
    collection=database["people"]
    return collection
def load(uri="localhost",datapath=r"C:\Users\eleves\Desktop\STAGE 2A\data.json.codechallenge.janv22.RHOBS.json"):
    coll=collection(uri=uri)
    with open(datapath,"r") as fp:
        data=json.load(fp)
        for person in data:
            coll.insert_one(person)
```

###  une fonction qui renvoie les entreprises de plus de N personnes.

```python
def company_avec_plus_de_n_personnes(n):
    # connection à la base de données MongoDB
    client = MongoClient()
    db = client.rhobs
    # Les différentes entreprises trouvées dans la base de données 
    companies = {}
    for person in db.people.find():
        company = person['company']
        if company in companies:
            companies[company] += 1
        else:
            companies[company] = 1

    result = [company for company in companies if companies[company] > n]
    return result
```

###   une fonction qui prend en paramètre un métier et qui renvoie la pyramide des âges pour ce métier.

Dans le fichier y a pas l'information sur l'âge qui est donné, on utilise du coup la bibliothèque datetime afin de calculer l'âge à partir de la date de naissance. 

```python
from datetime import datetime
from datetime import date

def ages_metier(metier):
    # connection à la base de données MongoDB
    client = MongoClient()
    db = client.rhobs
    # la liste des personnes avec le métier donné
    personnes = db.people.find({'job': metier})
    personnes = list(personnes)
    # calculer leurs ages  
    birthday = [datetime.strptime(p['birthdate'], "%Y-%m-%d").date() for p in personnes]
    today=date.today()
    ages=[today.year - b.year - ((today.month, today.day) < (b.month, b.day)) for b in birthday]
    pyramide = {'<25': 0, '25-35': 0, '35-45': 0, '45-55': 0, '55-65': 0, '+65':0}
    for age in ages:
        if age < 25:
            pyramide['<25'] += 1
        elif age < 36:
            pyramide['25-35'] += 1
        elif age < 46:
            pyramide['35-45'] += 1
        elif age < 56:
            pyramide['45-55'] += 1
        elif age < 66:
            pyramide['55-65'] += 1
        else:
            pyramide['+65'] += 1

    return pyramide
```

```python
ages_metier('expert automobile')
```
